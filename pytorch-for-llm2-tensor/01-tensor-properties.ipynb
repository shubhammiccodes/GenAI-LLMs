# Tensor Properties

## import requird pkg

```python
import numpy as np
import torch
```

### create rank 1 tensor without autograd

```python
# rank 1 tensor - vector
vector = torch.tensor([10,20,30,40,50],  dtype = torch.float32)
vector
```

```python
print(f"data type of each val inside the tensor: {vector.dtype}")
```

```python
print(f"size of tensor = {vector.numel()} or {vector.size()}")
```

```python
print(f"shape of tensor: {vector.shape}")
```

```python
print(f"no of dimension: {vector.ndim} or {vector.dim()}")
```

```python
print(f"device the vector is going to perform the operation: {vector.device}")
```

```python
print(f"is autograd  enabled: {vector.requires_grad}")
```

```python
print(f"is this tensor is a leaf node of DAG: {vector.is_leaf}")
```

### create rank2 tensor without autograd 

```python
# rank 3 tensor - matrix
matrix = torch.tensor([[10, 20],[30, 40]], dtype = torch.float32)
```

```python
print(f"data type of every value inside the tensor: {matrix.dtype}")
print(f"size of tensor: {matrix.numel()} or {matrix.size()}")
print(f"shape of tensor: {matrix.shape}")
print(f"no of dimensions: {matrix.dim()} or {matrix.ndim}")
print(f"device the vector is going to perform the operations: {matrix.device}")
print(f"is autograd enabled: {matrix.requires_grad}")
print(f"is this tensor a leaf node of DAG: {matrix.is_leaf}")
```
