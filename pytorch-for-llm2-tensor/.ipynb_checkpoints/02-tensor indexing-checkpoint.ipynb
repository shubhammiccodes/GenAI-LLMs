# Tensor Indexing

## import required pkg

```python
import numpy as np
import torch
```

### create rank1 tensor 

```python
vector = torch.tensor([10,20,30,40,50])
vector
```

### +ve Indexing

```python
print(f"vector[0] = {vector[0]}")
print(f"vector[1] = {vector[1]}")
print(f"vector[len(vector)-1 = {vector[len(vector)-1]}")
```

### -ve Indexing

```python
# print(f"vector[len(tensor) -1] = {vector[len(tesnor)-1]}")
print(f"vector[-1] = {vector[-1]}")
print(f"vector[-2] = {vector[-2]}")
print(f"vector[-len(vector)] = {vector[-len(vector)]}")
```

## create rank 2 tensor

```python
matrix = torch.tensor([[10,20], [30,40], [50,60]])
matrix
```

### +ve Indexing

```python
print(f"the first row (matrix[0]) = {matrix[0]}")
print(f"matrix[0][0] = {matrix[0][0]}")
```

```python
print(f"the last row (matrix[-1]) = {matrix[-1]}")
```

```python
print(f"matrix[1][1] = {matrix[1][1]}")
print(f"matrix[1,1] = {matrix[1,1]}")
```
