# Tensor  reducing operations


## import required pkg

```python
import numpy as np
import torch
```

## create rank1 tensor

```python
vector = torch.tensor([10, 20, 30, 40, 50, 60], dtype = torch.float32)
vector
```

## perform reducing (aggregation) operation

```python
print(f"max val = {vector.max()}")
```

```python
print(f"min val = {vector.min()}")
```

```python
print(f"mean val = {vector.mean()}")
```

```python
print(f"sum of all val = {vector.sum()}")
```

```python
print(f"product of all val = {vector.prod()}")
```

```python
print(f"standard deviation = {vector.std()}")
```

## create rank 2 tensor

```python
matrix =  torch.tensor([[10, 20, 30], [40, 50, 60]], dtype = torch.float32)
matrix
```

## perform reducing oper.

```python
print(f"max value = {matrix.max()}")
```

```python
print(f"min value = {matrix.min()}")
```

```python
print(f"mean value = {matrix.mean()}")
```

```python
print(f"sum of all the values = {matrix.sum()}")
```

```python
print(f"product of all the vlaues = {matrix.prod()}")
```

```python
print(f"standard deviation = {matrix.std()}")
```

### *perform oper. columnwise

```python
# dim = 0 : column
print(f"max val = {matrix.max(dim = 0)}")
print(f"max val = {matrix.max(dim = 0).values}")
```
