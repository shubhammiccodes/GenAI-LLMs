# Tensor Slicing

## import required pkg

```python
import numpy as np
import torch
```

### create rank 1 tensor 

```python
vector = torch.tensor([10,20,30,40,50,60])
vector
```

### slicing

```python
# start 2 and stop 5(excluded)
print(f"vector[2:5] = {vector[2:5]}")
```

```python
# start = 0 and stop = 4 (excluded)
print(f"vector[0:4] = {vector[0:4]}")

# start = 0 and stop = 4 (excluded)
print(f"vector[:4] = {vector[:4]}")
```

```python
# start = 3 and stop =  6(excluded)
print(f"vector[3:6] = {vector[3:6]}")

# start = 3 and stop =  6(excluded)
print(f"vector[3:] = {vector[3:]}")
```

```python
# get all the values
print(f"vector[0:6] ={vector[0:6]}")

print(f"vector[0:] = {vector[0:]}")

print(f"vector[:6] = {vector[:6]}")

print(f"vector[:] = {vector[:]}")
```

```python
# get all the alternate val
print(f"vector[::2] = {vector[::2]}")
```

```python
#  get all third val
print(f"vector[::3] = {vector[::3]}")
```

```python
# get all value in reverse order
reversed_vector = torch.flip(vector, dims = [0])
print(f"reversed = {reversed_vector}")

reversed_vector = torch.flip(vector, dims = [-1])
print(f"reversed = {reversed_vector}")
```

### create rank2 tensor

```python
matrix = torch.tensor([[10,20], [30,40], [50,60]])
matrix
```

```python
# get all  rows and cols
print(f"matrix[:] = {matrix[:]}")
print(f"matrix[:, :] = {matrix[:, :]}")
```

```python
# get first col
matrix[:, 0]
```

```python
# get 2nd col
matrix[:, 1]
```
