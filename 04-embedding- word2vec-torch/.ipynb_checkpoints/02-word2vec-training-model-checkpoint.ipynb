### import required package

```python
import numpy as np
from gensim.models import Word2Vec
```

```python
# input from training the word2vec model 
sentences = [
    ["i", "love", "ai"],
    ["i", "love", "machine", "learning"],
    ["ai", "is", "fun"],
    ["machine", "learning", "is", "powerful"],
    ["i", "enjoy", "learning"],
    ["deep", "learning", "uses", "neural", "networks"]
]
```

```python
# create 2 word2vec model
model = Word2Vec(
    sentences = sentences, # input
    vector_size = 10,      # dimension
    epochs = 100,          #no of epochs
    min_count = 1, 
    sg = 1
)
```

```python
# get the vector of required token
model.wv["i"]
```

```python
model.wv["learning"]
```

```python
# since the word python is NOT present in the wv lookup table, this line will generate an error
# model.wv["python"]
```

```python
model.wv.similarity("ai", "machine")
```
