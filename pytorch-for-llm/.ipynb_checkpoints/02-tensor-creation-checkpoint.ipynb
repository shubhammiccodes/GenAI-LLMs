# Tensor Creation

## import required pkg

```python
import numpy as np
import torch
```

### using python data

```python
# using python list
scalar = torch.tensor(10)
vector = torch.tensor([10,20,30,40,50])
matrx = torch.tensor([[10,20], [30,40]])
vector
```

```python
# using python tuple
vector_using_tuple = torch.tensor((10,20,30,40,50))
vector_using_tuple
```

### using numpy array

```python
array1 = np.array([10,20,30,40,50])
vector_using_np = torch.tensor(array1)
vector_using_np
```

```python
array2 = np.array([[10,20], [30,40]])
matrix_using_np = torch.tensor(array2)
matrix_using_np
```
