# tensor creating using  factory methods

### import required pkg

```python
import numpy as np
import torch
```

## empty tensor

```python
empty_tensor = torch.empty(2,2)
empty_tensor
```

```python
empty_tensor_copy = torch.empty_like(empty_tensor)
empty_tensor_copy
```

## zeros

```python
zeroed_tensor = torch.zeros(3, 3)
zeroed_tensor
```

## ones

```python
one_tensor = torch.ones(3, 3)
one_tensor
```

## full

```python
full_tensor = torch.full((3,3), 45)
full_tensor
```

## using Arange

```python
# the value will start from 0 and stop on 9
# value 10 will be excluded
sample_tensor = torch.arange(0, 10)
sample_tensor
```

## identity matrix

```python
identity_matrix = torch.eye(5)
identity_matrix
```

## diagonal matrix 

```python
t1 = torch.tensor([1,2,3,4,5,6,7])
t1
```

```python
diagonal_tensor = torch.diag(t1)
diagonal_tensor
```

## using random integers

```python
ran_int_tensor = torch.randint(0, 20, (5, 5))
ran_int_tensor
```

## using random permutation

```python
ran_per_tensor = torch.randperm(10)
ran_per_tensor
```
