# Vector Tesnor(Rank1)

```python
# list of nos
number_list = [10,20,30,40,50]
print(f"number_list = {number_list}, type = {type(number_list)}")
```

```python
# numpy array of numbers
numbers_array = np.array([10,20,30,40,50])
print(f"numbers_tensor = {numbers_array}, type = {type(numbers_array)}")
print(f"numbers of dimension = {numbers_array.ndim}")
```

```python
# vector of numbers
numbers_vector = torch.tensor([10,20,30,40,50])
print(f"numbers_vector = {numbers_vector}, type = {type(numbers_vector)}")
print(f"numbers of dimension = {numbers_vector.ndim}")
```

## matrix Tensor (Rank -2)

```python
matrix_tensor = torch.tensor([[10,20], [30,40], [50,60]])
print(f"matrix_tensor = {matrix_tensor}, type = {type(matrix_tensor)}")
print(f"numbers of dimension = {matrix_tensor.ndim}")
```

## Matrix Tesnor(Rank-3)

```python
matrix_tensor_3d = torch.tensor([[[10,20], [30,40]], [[50,60], [70,80]]])
print(f"matrix_tensor_3d = {matrix_tensor_3d}, type= {type(matrix_tensor_3d)}")
print(f"matrix_tensor_3d_dim = {matrix_tensor_3d.ndim}")
```
